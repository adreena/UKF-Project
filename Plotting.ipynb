```python
import matplotlib.pyplot as plt
import csv
```

```python
def draw_data(file_name, title):
    frame= []
    nis = []
    e_top = []
    e_low = []
    values_in_between=0
   
    with open(file_name,'r') as csvfile:
    #     plots = csv.reader(csvfile, delimiter=',')
        lis=[line.split() for line in csvfile][1:]
        count = 0
        for line in lis:
            row = line[0].split(',')
            if len(row) == 2 and row[-1]!='':
                frame.append(row[0])
                nis.append(round(float(row[1]), 2))
                e_top.append(7.85)
                e_low.append(0.352)
                if float(row[1])< 7.85 and float(row[1]) > 0.352:
                    values_in_between+=1
    plt.plot(frame,nis,'r',label='NIS') # plotting NIS  
    plt.plot(frame,e_top,'g',label='7.85')
    plt.plot(frame,e_low,'b',label='0.352')
    plt.xlabel('frame')
    plt.ylabel('NIS')
    plt.legend()
    plt.title(title)
    plt.show()
    plt.close()
    print(100* values_in_between/len(e_low))
```

```python
# 30 , 30
draw_data('nis_output_file.csv', 'NIS')
```
