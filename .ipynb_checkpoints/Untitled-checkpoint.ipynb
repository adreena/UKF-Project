```python
import matplotlib.pyplot as plt
import csv
```

```python
def draw_data(file_name, title):
    frame= []
    nis = []
    e_ = []
    
    with open(file_name,'r') as csvfile:
    #     plots = csv.reader(csvfile, delimiter=',')
        lis=[line.split() for line in csvfile][1:]
        count = 0
        for line in lis:
            row = line[0].split(',')
            if len(row) == 2 and row[-1]!='':
                frame.append(row[0])
                nis.append(round(float(row[1]), 2))
                e_.append(7.85)
    plt.plot(frame,nis,'r',label='NIS') # plotting NIS  
    plt.plot(frame,e_,'g',label='NIS')
    plt.xlabel('frame')
    plt.ylabel('NIS')
    plt.legend()
    plt.title(title)
    plt.show()
    plt.close()
```

```python
# 30 , 30
draw_data('nis_output_file.csv', 'NIS')
```
